# Autocorreletion<a class="anchor" id="aut"></a>
Here the steps are 50k with 50% ratio of acceptance.\
The formula used for the auto correlation is given by
$$
\chi(t) =\frac{\frac{1}{t_{max}-t}\sum_0^{t_{max}-t} m(t')m(t'+t)-\frac{1}{t_{max}-t}\sum_0^{t_{max}-t}m(t')\times\frac{1}{t_{max}-t}\sum_0^{t_{max}-t}m(t'+t)}{\frac{1}{t_{max}}\sum_0^{t_{max}} m^2(t')-\left(\frac{1}{t_{max}}\sum_0^{t_{max}} m(t')\right)^2}
$$
And it's expected an exponential decay of the autocorrelation

```python
epotential = pd.read_csv("MD_MC/MDNVE_MCNVT/MC/output_epot.dat",header=None)

l= int(len(epotential)/3)
x = np.linspace(0,l,l)
# in totale devo avere 5*10^5 punti su n blochi etc. io ne uso 60
M = l

x = np.arange(M)+1

colors=["r","b","g"]
state = ["Gas","Liquid","Solid"]
plt.figure(figsize=(10,6))
for i in range(3):
    plt.plot(x,epotential[1][i*l:(i+1)*l],color=colors[i],label=state[i])
plt.legend()
```

```python
tmax = M
taumax = 200

plt.figure(figsize=(8,6))

for i in range(3):

    chi = []

    #print("\ntau (nth of "+str(taumax)+") - "+phase[i]+": ")
    for tau in range(taumax):
        s1=0
        s2=0
        s3=0
        s4=0
        s5=0

        # autocorrelation: see lect.6, p.31
        for tp in range(tmax-tau):
            s1 += epotential[1][tp+i*l]*epotential[1][tp+tau+i*l]
            s2 += epotential[1][tp+i*l]
            s3 += epotential[1][tp+tau+i*l]

        for tp in range(tmax):
            s4 += epotential[1][tp+i*l]*epotential[1][tp+i*l]
            s5 += epotential[1][tp+i*l]

        fatt = 1/(tmax-tau)
        fatt2 = 1/tmax

        num = fatt*s1-fatt*s2*fatt*s3
        den = fatt2*s4 - (fatt2*s5)*(fatt2*s5)

        chi.append(num/den)

        if tau%10==0:
            print(tau, end= ", ")

    plt.plot(chi, color = colors[i],label=state[i])
    plt.xlabel("Monte carlo time")#Distance between different measurments
    plt.ylabel("Autocorrelation")
    plt.legend()
    plt.grid()

plt.title("Autocorrelation vs time distance", fontsize = 18, fontweight = "bold")
plt.show
```

# Statistical error<a class="anchor" id="se"></a>
To evaluate the statistical error I've used 50k steps in 10 blocks

```python
L_min = 10
L_max = 5000
#L = M/N
def Error(ave, ave2, n):
    if(n==0): 
        return 0
    return np.sqrt((ave2-ave**2)/n)

fig=plt.figure(figsize=(17, 17))

for i in range(3):
    plt.subplot(3,1,i+1)
    plt.title("Statistical error - "+state[i], size='xx-large', weight='bold')

    y=epotential[1] 
    erro=[]
    x=[]

    print("\nL (of "+str(L_max)+") - "+state[i]+": ")
    for L in range(L_min,L_max, 10): 
        if L%500==0:
            print(L, end = ", ")
        N = int(M/L)
        mean=0
        mean2=0
        error=0       
        for n in range(N):
            mean_block=0
            for l in range(L):
                mean_block+=epotential[1][(l+n*L)+i*M]
            mean_block/=L
            mean=(mean*n+mean_block)/(n+1)
            mean2=(mean2*n+mean_block**2)/(n+1)
            err=Error(mean, mean2,n)
            
        erro.append(err)
        x.append(L)

    plt.plot(x, erro, label = state[i], color = colors[i])
        
    plt.ylabel("Error")
    plt.xlabel('L')
    plt.grid(True)

plt.tight_layout()
plt.show()

```

The figures show that the statistical error tends to stabilize and oscillate along a constant value.

# Radial distribution function<a class="anchor" id="rdf"></a>
The rdf g(r) describes the probability to find a particle ad a distance $r$ to another, hence $r\in[0,L/2]$. In the following I show the mean and error of the radial distribution for gas,liquid and solid.

```python
gdr = pd.read_csv("MD_MC/MDNVE_MCNVT/output_gdr_MC.dat",header=None)
gdrf = pd.read_csv("MD_MC/MDNVE_MCNVT/MC/output_gdr_MD.dat",header=None)

l = int(len(gdr)/3)
d = int(len(gdrf)/3)


dist = np.arange(0,99)

# box scaling
rho = [0.05, 0.8, 1.1]
npart = 108
L = [0.0,0.0,0.0]
color = ["r","b","g"]
label = ["Gas","Liquid","Solid"]
for i in range(3):
    L[i] = ((rho[i]/npart)**(-1/3))/200
    plt.errorbar(dist*L[i],gdr[i*l:(i+1)*l].mean(),yerr=gdr[i*l:(i+1)*l].std(),color=color[i],label=label[i])
    plt.xlim(0.0,2.5)
    plt.xlabel("Distance")
    plt.ylabel("g(r)")
```
