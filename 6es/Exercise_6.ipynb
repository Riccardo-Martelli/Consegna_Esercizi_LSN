# Exercise 6
The Ising model in 1D is very useful because it has an analytical solution, therefore one can make a simulation of it and compare it with the theoretical result.\
The sampling methods used are Metropolis and Gibbs, these two methods are then compared.\
The system is analyzed when a magnetic field h is h=0 and h=0.02 with a temperature between 0.5 and 3 Kelvin.\
Before proceeding with the simulation an equilibration has been done.\
The errors have been plotted separately because the error bar where too short to be seen in a plot. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

plt.rcParams.update({'font.size': 18})

ene_metr = pd.read_csv('Ising_1D/ISING_1D/output.ene.0',header=None)
heat_metr = pd.read_csv('Ising_1D/ISING_1D/output.heat.0',header=None)
chi_metr = pd.read_csv('Ising_1D/ISING_1D/output.chi.0',header=None)

ene = pd.read_csv('Ising_1D/ISING_1D/ene.0',header=None)
heat = pd.read_csv('Ising_1D/ISING_1D/heat.0',header=None)
chi = pd.read_csv('Ising_1D/ISING_1D/chi.0',header=None)
mag_metr = pd.read_csv('Ising_1D/ISING_1D/output.mag.0',header=None)
mag2_metr = pd.read_csv('Ising_1D/ISING_1D/mag.0.020',header=None)

equil = pd.read_csv('Ising_1D/ISING_1D/output.eq',header=None)

#x = np.linspace(0.5,1.5,int(len(ene[1])/2))
#y = np.linspace(0.5,1.5,int(len(ene_metr[1])))
l= int(len(ene)/2)

z = ene.groupby([0])[0].mean()

N=2000
x= np.linspace(0,N,N)
ran = int(len(equil)/(2))
plt.figure(figsize=(8,6))
plt.title("Equilibration U/N  Metropolis, h=0")
for i in range(5):
     plt.plot(x,equil[1][i*N:(i+1)*N],label="T={}".format(0.5+i*0.25))
plt.legend()
plt.xlabel("Steps")
plt.ylabel("U/N")
```

```python
plt.figure(figsize=(8,6))
plt.title("Equilibration U/N Gibbs, h=0")
for i in range(5):
     plt.plot(x,equil[1][ran +i*N:ran + (i+1)*N],label="T={}".format(0.5+i*0.25))
plt.legend()
plt.xlabel("Steps")
plt.ylabel("U/N")
```

```python
points=100
T = np.linspace(0.5,3.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*th*( 1 + thN/(th*th) )/( 1 + thN )
plt.plot(T, e,label="U/N analytic")
plt.title('Ising 1D, internal energy')
plt.xlabel('T')
plt.ylabel('U/N')

plt.plot(z[0:],ene[:l].groupby([0])[1].mean(),linestyle="none",marker="o",label="Metro")
plt.plot(z[0:],ene[l:].groupby([0])[1].mean(),linestyle="none",marker="*",label="Gibbs")
plt.legend()
plt.show()

plt.title("Progression of the errors for U")
plt.plot(ene[:l].groupby([0])[0].mean(),ene[:l].groupby([0])[2].mean(),label="Metro")
plt.plot(ene[l:].groupby([0])[0].mean(),ene[l:].groupby([0])[2].mean(),label="Gibbs")
plt.xlabel("T")
plt.ylabel("Error")
plt.legend()

plt.show()
```

Since the errorbar is too small the relitie errors are shown in another picture

```python
heatt=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
plt.plot(T, heatt, label = "Cap term analytic")
plt.title('Ising 1D, Thermal capacity')
plt.xlabel('T')
plt.ylabel('C')
plt.plot(z,heat[:l].groupby([0])[1].mean(),linestyle="none",marker="o",label="Metro")
plt.plot(z,heat[l:].groupby([0])[1].mean(),linestyle="none",marker="*",label="Gibbs")
plt.legend()

plt.show()

plt.title("Progression of the errors for C")
plt.plot(heat[:l].groupby([0])[0].mean(),heat[:l].groupby([0])[2].mean(),label="Metro")
plt.plot(heat[l:].groupby([0])[0].mean(),heat[l:].groupby([0])[2].mean(),label="Gibbs")
plt.xlabel("T")
plt.ylabel("Error")
plt.legend()

plt.show()
```

```python
h=0.02 #external field
b = 1/T

plt.title('Ising 1D, Magnetization with h=0.02')

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
plt.plot(T, M, label = "M analytic ")
plt.plot(mag2_metr[:l].groupby([0])[1].mean(),linestyle="none",marker="o",label="Metro")
plt.plot(mag2_metr[l:].groupby([0])[1].mean(),linestyle="none",marker="*",label="Gibbs")
plt.xlabel('T')
plt.ylabel('$M$')
plt.legend()

plt.show()

plt.title("Progression of the errors for M")
plt.plot(mag2_metr[:l].groupby([0])[0].mean(),mag2_metr[:l].groupby([0])[2].mean(),label="Metro")
plt.plot(mag2_metr[l:].groupby([0])[0].mean(),mag2_metr[l:].groupby([0])[2].mean(),label="Gibbs")
plt.xlabel("T")
plt.ylabel("Error")
plt.legend()

plt.show()
```

```python
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
plt.plot(T, X, label = "$\chi$ analytic")
plt.title('Ising 1D, Magnetic susceptibility')
plt.ylabel('$\chi$')
plt.xlabel('T')
plt.plot(chi[:l].groupby([0])[1].mean(),linestyle="none",marker="o",label="Metro")
plt.plot(chi[l:].groupby([0])[1].mean(),linestyle="none",marker="*",label="Gibbs")
plt.legend()

plt.show()

plt.title("Progression of the errors for $\chi$")
plt.plot(chi[:l].groupby([0])[0].mean(),chi[:l].groupby([0])[2].mean(),label="Metro")
plt.plot(chi[l:].groupby([0])[0].mean(),chi[l:].groupby([0])[2].mean(),label="Gibbs")
plt.xlabel("T")
plt.ylabel("Error")
plt.legend()

plt.show()
```
