# Exercise 4

Here we consider a micro canonical ensemble of 108 particles and want to calculate some termodynamic properties, total (internal) energy per particle, $E/N$, the potential energy per particle, $U/N$, the kinetic energy per particle, $K/N$, the temperature, $T$ and the pressure P. The particle interact via the Lennard-Jones potential, and the algorithm used is the Verlet algorithm.

Before starting the simulation there is the need to equilibrate the temperature, indeed by setting a initial value for the temperature T, after the system evolves, it reaches a constant value $T^*$ and that is the value we want.
I've found the following initial values for T
- T = 0.96 for the gas phase
- T = 1.99 for the liquid phase
- T = 1.54 for the solid phase\
in order to reach the requested values $T^*$.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

plt.rcParams.update({'font.size': 20})

pres = pd.read_csv("./MD_MC/MDNVE_MCNVT/output_press.dat",header = None)
epot = pd.read_csv("./MD_MC/MDNVE_MCNVT/output_epot.dat",header = None)
ekin = pd.read_csv("./MD_MC/MDNVE_MCNVT/output_ekin.dat",header = None)
etot = pd.read_csv("./MD_MC/MDNVE_MCNVT/output_etot.dat",header = None)
temp = pd.read_csv("./MD_MC/MDNVE_MCNVT/output_temp.dat",header = None)
equil_temp = pd.read_csv("./MD_MC/MDNVE_MCNVT/temp_equilibration.dat",header = None)

x = np.arange(len(pres[1]))
x=x[:60]

y=np.arange(len(equil_temp)/3)
N= 30000
```

# Equilibration of T
In the following graphs it can be seen that the temperature in the gas phase is less stable, while the liquid and solid phase oscillate around $T^*$.\
To calibrate I've used one block with 30k steps.

```python
plt.figure(figsize=(18,12))

plt.suptitle("Equilibration time for the three states",fontsize=26,fontweight='bold')

plt.subplot(311)
plt.title("Gas",fontweight='bold')
plt.axhline(1.2,color='r',label="$T^* GAS= 1.2$")
plt.plot(y,equil_temp[:N],label="Temperature")
plt.xlabel("steps")
plt.ylabel("T")
plt.legend()

plt.subplot(312)
plt.title("Liquid",fontweight='bold')
plt.axhline(1.1,color='r',label="$T^* LIQUID = 1.1$")
plt.plot(y,equil_temp[N:2*N],label="Temperature")
plt.xlabel("steps")
plt.ylabel("T")
plt.legend()

plt.subplot(313)
plt.title("Solid",fontweight='bold')
plt.axhline(0.8,color='r',label="$T^* SOLID = 0.8$")
plt.plot(y,equil_temp[2*N:3*N],label="Temperature")
plt.xlabel("steps")
plt.ylabel("T")
plt.legend()

plt.tight_layout()

plt.show()
```

## Thermodynamic properties
In the following I calculate the thermodynamic said before to do so I've used 60 block with 2k steps each.

## Gas

```python
plt.figure(figsize=(20,12))

plt.suptitle("GAS",fontsize=26,fontweight='bold')

plt.subplot(2,3,1)
plt.title("Pressione",fontweight='bold')
plt.errorbar(x,pres[1][:60],yerr=pres[3][:60],ecolor='r')
plt.xlabel("Blocks")
plt.ylabel("Pressione")
plt.grid(True)

plt.subplot(2,3,2)
plt.title("Temperatura",fontweight='bold')
plt.errorbar(x,temp[1][:60],yerr=temp[3][:60],ecolor='r')
plt.axhline(1.2,color='g',label="$T^*$")
plt.xlabel("Blocks")
plt.ylabel("$T^*$")
plt.legend()
plt.grid(True)

plt.subplot(2,3,3)
plt.title("Energia potenziale",fontweight='bold')
plt.errorbar(x,epot[1][:60],yerr=epot[3][:60],ecolor='r')
plt.xlabel("Blocks")
plt.ylabel("$U^*/N$")
plt.grid(True)

plt.subplot(2,3,4)
plt.title("Energia cinetica",fontweight='bold')
plt.errorbar(x,ekin[1][:60],yerr=ekin[3][:60],ecolor='r')
plt.xlabel("Blocks")
plt.ylabel("$K^*/N$")
plt.grid(True)

plt.subplot(2,3,5)
plt.title("Energia totale",fontweight='bold')
plt.errorbar(x,etot[1][:60],yerr=etot[3][:60],ecolor='r')
plt.xlabel("Blocks")
plt.ylabel("$E^*/N$")
plt.grid(True)

plt.tight_layout()

plt.show()
```

## liquid

```python
plt.figure(figsize=(20,12))

plt.suptitle("LIQUIDO",fontsize=26,fontweight='bold')

plt.subplot(2,3,1)
plt.title("Pressione",fontweight='bold')
plt.errorbar(x,pres[1][60:120],yerr=pres[3][60:120],ecolor='r')
plt.xlabel("Blocks")
plt.ylabel("Pressione")
plt.grid(True)

plt.subplot(2,3,2)
plt.title("Temperatura",fontweight='bold')
plt.errorbar(x,temp[1][60:120],yerr=temp[3][60:120],ecolor='r')
plt.axhline(1.1,color='g',label="$T^*$")
plt.xlabel("Blocks")
plt.ylabel("$T^*$")
plt.legend()
plt.grid(True)

plt.subplot(2,3,3)
plt.title("Energia potenziale",fontweight='bold')
plt.errorbar(x,epot[1][60:120],yerr=epot[3][60:120],ecolor='r')
plt.xlabel("Blocks")
plt.ylabel("$U^*/N$")
plt.grid(True)

plt.subplot(2,3,4)
plt.title("Energia cinetica",fontweight='bold')
plt.errorbar(x,ekin[1][60:120],yerr=ekin[3][60:120],ecolor='r')
plt.xlabel("Blocks")
plt.ylabel("$K^*/N$")
plt.grid(True)

plt.subplot(2,3,5)
plt.title("Energia totale",fontweight='bold')
plt.errorbar(x,etot[1][60:120],yerr=etot[3][60:120],ecolor='r')
plt.xlabel("Blocks")
plt.ylabel("$E^*/N$")
plt.grid(True)

plt.tight_layout()

plt.show()
```
