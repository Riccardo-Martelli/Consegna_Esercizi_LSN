# Exercise 5
The aim of this exercise is to analyze the wave function of the hydrogen atom in particular the ground state and the first exited state, from these two function the
expectation values for the radius turns out to be exactly:
$$
\left\langle r \right\rangle_{\Psi_{1,0,0}} = \frac{3}{2}a_0
\quad
\left\langle r \right\rangle_{\Psi_{2,1,0}} = 5 a_0
$$
To check these results the Metropolis algorithm is implemented and two different distributions are used to propose a step value, namely the uniform and gaussian distribution.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

plt.rcParams.update({'font.size': 18})

points = pd.read_csv("points_psi_one_unif.txt")
points2 = pd.read_csv("points_psi_two_unif.txt")
sums = pd.read_csv("avg_psi_one_unif.txt")
sums2 = pd.read_csv("avg_psi_two_unif.txt")

x = np.linspace(1,100,len(sums["mean"]))
y = np.linspace(1,100,len(sums2["mean"]))
```

# Equilibration/number of steps <a class="anchor" id="eq"></a>

Before jumping in the simulation is necessary to equilibrate the system, i.e. we don't want the electron to start from unlikely positions. This is helpful because it also gives a nice view of how much time it takes for the system to equilibrate. Here the equilibration for uniform and gaussian method are show.

```python
uequ = pd.read_csv("unif_equilibration.txt")
gequ = pd.read_csv("gaus_equilibration.txt")

l= np.linspace(1,len(gequ["mean"]),len(gequ["mean"]))

plt.title("Equilibration average r for $\psi_{1,0,0}$")
plt.axhline(1.5,color='g')
plt.errorbar(l,uequ['mean'],yerr=uequ["err"],ecolor='r',label="Uniform")
plt.xlabel("Blocks")
plt.ylabel("$<r>/a_0$")
plt.legend()
plt.show()

plt.title("Equilibration average r for $\psi_{1,0,0}$")
plt.axhline(1.5,color='g')
plt.errorbar(l,gequ['mean'],yerr=gequ["err"],ecolor='r',label="Gauss")
plt.xlabel("Blocks")
plt.ylabel("$<r>/a_0$")
plt.legend()
plt.show()
```

It's easy to see that the gauss method has a nicer convergence than the uniform one.\
In both cases the convergence appears after around 10k steps, therefore in the following I've used 100 blocks of $10^5$ steps.

# Uniform <a class="anchor" id="unif"></a>

Here I present, both ground and exited state using the uniform method, the staring positions are (2,0,0) for the ground state and (5,0,0) for the exited state.

```python
plt.title("Average values of r for $\psi_{1,0,0}$")
plt.axhline(1.5,color='g')
plt.errorbar(x,sums['mean'],yerr=sums["err"],ecolor='r')
plt.xlabel("Blocks")
plt.ylabel("$<r>/a_0$")
plt.show()
```

```python
plt.title("Average values of r for $\psi_{2,1,0}$")
plt.axhline(5,color='g')
plt.errorbar(y,sums2["mean"],yerr=sums2["err"],ecolor='r')
plt.xlabel("Blocks")
plt.ylabel("$<r>/a_0$")
```

```python
x = points["x"]
y = points["y"]
z = points["z"]

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.set_title("Distribution $\psi_{1,0,0}$")
ax.scatter(x, y, z, c=z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 55)
```
