# Exercise 10
In this exercise I've used the MPI library to parallelize the travel salesman genetic algorithm using american capitals instead.I've used 4 cores and a population of 15000 for 15000 generations.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

plt.rcParams.update({'font.size': 18})

cities = pd.read_csv("American_capitals.dat",sep='\t',names=["x","y"])
```

# Continent with parallelization not exchanging the best individuals

```python
migr0 = pd.read_csv("data_continent/cromosomes_0",header=None)
migr1 = pd.read_csv("data_continent/cromosomes_1",header=None)
migr2 = pd.read_csv("data_continent/cromosomes_2",header=None)
migr3 = pd.read_csv("data_continent/cromosomes_3",header=None)

L2_best_migr0 = pd.read_csv("data_continent/L2_best_0",names=["L2","gen"])
L2_best_migr1 = pd.read_csv("data_continent/L2_best_1",names=["L2","gen"])
L2_best_migr2 = pd.read_csv("data_continent/L2_best_2",names=["L2","gen"])
L2_best_migr3 = pd.read_csv("data_continent/L2_best_3",names=["L2","gen"])
```

```python
L2_best_migr0["L2"].min()
best = migr0.iloc[L2_best_migr0["L2"].argmin()].to_numpy()
best = best -1

plt.figure(figsize=(6,6))
plt.title("Best path core 0")
plt.plot(cities["x"],cities["y"],color="r",linestyle="",marker="o",label = "Cities")
plt.plot(cities["x"][0],cities["y"][0],marker="*", markersize = 7,label = "Start", linestyle = "")
plt.plot(cities["x"][best],cities["y"][best],color="g",label="Path")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.90))
plt.xlabel("x")
plt.ylabel("y")

L2_best_migr1["L2"].min()
best = migr1.iloc[L2_best_migr1["L2"].argmin()].to_numpy()
best = best -1


plt.figure(figsize=(6,6))
plt.title("Best path core 1")
plt.plot(cities["x"],cities["y"],color="r",linestyle="",marker="o",label = "Cities")
plt.plot(cities["x"][0],cities["y"][0],marker="*", markersize = 7,label = "Start", linestyle = "")
plt.plot(cities["x"][best],cities["y"][best],color="g",label="Path")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.90))
plt.xlabel("x")
plt.ylabel("y")
```

```python
L2_best_migr2["L2"].min()
best = migr2.iloc[L2_best_migr2["L2"].argmin()].to_numpy()
best = best -1

plt.figure(figsize=(6,6))
plt.title("Best path core 2")
plt.plot(cities["x"],cities["y"],color="r",linestyle="",marker="o",label = "Cities")
plt.plot(cities["x"][0],cities["y"][0],marker="*", markersize = 7,label = "Start", linestyle = "")
plt.plot(cities["x"][best],cities["y"][best],color="g",label="Path")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.90))
plt.xlabel("x")
plt.ylabel("y")

L2_best_migr3["L2"].min()
best = migr3.iloc[L2_best_migr3["L2"].argmin()].to_numpy()
best = best -1


plt.figure(figsize=(6,6))
plt.title("Best path core 3")
plt.plot(cities["x"],cities["y"],color="r",linestyle="",marker="o",label = "Cities")
plt.plot(cities["x"][0],cities["y"][0],marker="*", markersize = 7,label = "Start", linestyle = "")
plt.plot(cities["x"][best],cities["y"][best],color="g",label="Path")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.90))
plt.xlabel("x")
plt.ylabel("y")
```

All 4 paths show similar results in finding the best path, the time is similar to a simulation without parallelization but the program produces four simulations instead of one.
