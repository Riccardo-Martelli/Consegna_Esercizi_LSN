```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import scipy.stats as stats

plt.rcParams.update({'font.size': 18})

# Dataframes per i vari lanci

df = pd.read_csv("Integral.txt")
df_2 = pd.read_csv("Integral_2.txt")
#df_er = pd.read_csv("err.txt")

M=100000              # Total number of throws
N=100                 # Number of blocks
L=int(M/N)            # Number of throws in each block, please use for M a multiple of N

x = np.arange(len(df["I_1"]))
#x*=L
```

```python
plt.axhline(1.0,color ='g')
plt.errorbar(x,df["I_1"],yerr=df["err_1"],ecolor='r')
plt.title("Calcolo di I con sampling uniforme ")
plt.xlabel(" Blocchi ")
plt.ylabel("$I$")
plt.grid(True)
```
