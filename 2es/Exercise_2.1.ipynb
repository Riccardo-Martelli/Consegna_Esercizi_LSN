# Exercise 2.1

In this exercise we want to calculate the integral 
$$I = \frac{\pi}{2}\int_0^1\cos\left(\frac{\pi}{2}x\right) \,dx =1$$
We will use two methods: uniform sampling and importance sampling \
The distribution used in the importance sampling a a liner distribution of the form $2(1-x)$

## Uniform sampling
In this case I've used a 100 blocks with 1000000 throws in each block.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import scipy.stats as stats

plt.rcParams.update({'font.size': 18})

# Dataframes per i vari lanci

df = pd.read_csv("2.1/Integral.txt")
df_2 = pd.read_csv("2.1/Integral_2.txt")
#df_er = pd.read_csv("err.txt")

M=100000              # Total number of throws
N=100                 # Number of blocks
L=int(M/N)            # Number of throws in each block, please use for M a multiple of N

x = np.arange(len(df["I_1"]))
#x*=L
```

```python
plt.axhline(1.0,color ='g')
plt.errorbar(x,df["I_1"],yerr=df["err_1"],ecolor='r')
plt.title("Values of I with uniform sampling ")
plt.xlabel(" Blocks ")
plt.ylabel("$I$")
plt.grid(True)
```

## Importance sampling
The number of blocks is the same as before. Here the linear distribution is sampled with the invesere CDF method.

```python
plt.axhline(1.0,color ='g')
plt.errorbar(x,df_2["I_2"],yerr=df_2["err_2"],ecolor='r')
plt.title("Values of I with importance sampling")
plt.xlabel("Blocks")
plt.ylabel("$I$")
plt.grid(True)
```

It can be easily noticed that the simulation with importance sampling is way more precice that the one with uniform sampling
