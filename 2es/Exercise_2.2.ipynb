We wanto to simulate a RW on a cubic lattice with adiacent distance of a=1. To do so we use $10^4$RWs of 100 step each. We want to measure $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ and we expect a diffusive behaviour proportional to $\sqrt{N}$.

## Cubic lattice
in this case at each discrete time the walker makes a forward or backward step of length equal to $a$ in one of the 3 principal directions of the lattice: $x$, $y$ or $z$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import scipy.stats as stats
from scipy.optimize import curve_fit

plt.rcParams.update({'font.size': 18})

# Dataframes per i vari lanci

df = pd.read_csv("2.2/Random_Walks_d.txt")
dfc = pd.read_csv("2.2/Random_Walks_c.txt")


M=10000              # Total number of throws
N=100                 # Number of blocks
L=int(M/N)            # Number of throws in each block, please use for M a multiple of N

x = np.arange(len(df["RW"]))
#x*=L


def fun(x,k):
    return k*np.sqrt(x)
fit =curve_fit(fun,x,df["RW"])

plt.title("RW on a cubic lattice with fit")
plt.errorbar(x,df["RW"],yerr=df["err"],ecolor='r')
plt.xlabel("Steps")
plt.ylabel("$\sqrt{< |r_N|^2 >_{RW}}$")
plt.plot(x,fit[0]*np.sqrt(x), label="k$\sqrt{N}$\n k=" +str(round(fit[0][0],5)))
plt.legend()
print("k=",fit[0])
```

## Continuum
In this case at each discrete time the walker makes a step of length equal to $a(=1)$ along a random direction obtained by sampling uniformly the solid angle: $\theta \in [0,\pi]$ and $\phi \in [0,2\pi]$

```python
fit =curve_fit(fun,x,dfc["RW"])

plt.title("RW continuum with fit")
plt.errorbar(x,dfc["RW"],yerr=dfc["err"],ecolor='r')
plt.xlabel("Steps")
plt.ylabel("$\sqrt{< |r_N|^2 >_{RW}}$")
plt.plot(x,fit[0]*np.sqrt(x), label="k$\sqrt{N}$\n k=" +str(round(fit[0][0],5)))
plt.legend()
print("k=",fit[0])
```

From the graphs it can be seen that the fit with $k\sqrt{N}$ is quite good.
