```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import scipy.stats as stats
from scipy.optimize import curve_fit

plt.rcParams.update({'font.size': 18})

# Dataframes per i vari lanci

df = pd.read_csv("Random_Walks_d.txt")
dfc = pd.read_csv("Random_Walks_c.txt")


M=10000              # Total number of throws
N=100                 # Number of blocks
L=int(M/N)            # Number of throws in each block, please use for M a multiple of N

x = np.arange(len(df["RW"]))
#x*=L


def fun(x,k):
    return k*np.sqrt(x)
fit =curve_fit(fun,x,df["RW"])

plt.title("RW discreto con fit")
plt.errorbar(x,df["RW"],yerr=df["err"],ecolor='r')
plt.xlabel("Steps")
plt.ylabel("$\sqrt{< |r_N|^2 >_{RW}}$")
plt.plot(x,fit[0]*np.sqrt(x), label="k$\sqrt{N}$\n k=" +str(round(fit[0][0],5)))
plt.legend()
print("k=",fit[0])
```

```python
fit =curve_fit(fun,x,dfc["RW"])

plt.title("RW continuo con fit")
plt.errorbar(x,dfc["RW"],yerr=dfc["err"],ecolor='r')
plt.xlabel("i")
plt.ylabel("$\sqrt{< |r_N|^2 >_{RW}}$")
plt.plot(x,fit[0]*np.sqrt(x), label="k$\sqrt{N}$\n k=" +str(round(fit[0][0],5)))
plt.legend()
print("k=",fit[0])
```
