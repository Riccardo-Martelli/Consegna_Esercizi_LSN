# Exercise 3

In this exercise I need to find Call and Put of the European option in order to verify the Black-Scholes equations.  
- European Call-option price:
$$ C\left[ S(t),t \right] = S(t) N(d_1) -K e^{-r(T-t)} N(d_2) $$
- European Put-option price:
$$ P\left[ S(t),t \right] = S(t) \left[ N(d_1)-1\right] -K e^{-r(T-t)} \left[ N(d_2)-1\right]$$

This theory assumes that the time evolution of the price of an asset has to be random and shows a behaviour analogous to the GBM, with interest rate risk-free constat that is analogous to the brownian drift and volatility $\sigma$.\
In both cases I have to de-exponentialize in order to calculate C and P, after calculating the final price S(T), hence
$$C[S(0),0] = e^{-rT}\max(0,S-K) $$
$$P[S(0),0] = e^{-rT}\max(0,K-S) $$
The pameters used are:
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams.update({'font.size': 18})

data = pd.read_csv('res.txt')
err = pd.read_csv('err.txt') 

x=np.arange(100)+1

C = [ 14.975790778311286  for i in range(len(x))]
P = [ 5.4595325819072364  for i in range(len(x))]
```

## Progressive averages

## Direct
In this case I've directly sampled directly the final asset price. The number of blocks is 100 and the number of samples is $10^4$

```python
plt.title("P one step")
plt.errorbar(x,data["P_one"],yerr=err["err_P_one"],ecolor='r',label="predicted values")
plt.plot(x,P,label="target value",color = 'black')
plt.xlabel("Blocks")
plt.ylabel("P(T)")
plt.grid(True)
plt.show()
```

```python
plt.title("C one step")
plt.errorbar(x,data["C_one"],yerr=err["err_C_one"],ecolor='r',label="predicted values")
plt.plot(x,C,label="target value",color = 'black')
plt.xlabel("Blocks")
plt.ylabel("C(T)")
plt.grid(True)
```

## Discretize
In this case I've used dicrete time values between [0,T], in order to achieve S(T). Do do so I've divided the interval in 100 points. The number of blocks etc. is the same as before

```python
plt.title("P discrete")
plt.errorbar(x,data["P_discr"],yerr=err["err_P_discr"],ecolor='r',label="predicted values")
plt.plot(x,P,label="target value",color = 'black')
plt.xlabel("Blocks")
plt.ylabel("P(T)")
plt.legend()
plt.grid(True)
plt.show()

plt.title("C discrete")
plt.errorbar(x,data["C_discr"],yerr=err["err_C_discr"],ecolor='r',label="predicted value")
plt.plot(x,C,label="target value",color = 'black')
plt.xlabel("Blocks")
plt.ylabel("C(T)")
plt.legend()
plt.grid(True)
```
