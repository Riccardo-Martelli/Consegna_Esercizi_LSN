# Exercise 9
The aim of this exercise is to find the minimum path length for a traveling sales man using genetic algorithm in two cases: 34 cities positioned on a circle and 34 cities inside a square.\
For the simulation I've used 50000 generation with a population of 10000 individuals, since the cities are 34 the number of genes is 34. The loss/cost function that I've used is the $L^2$ function.

## Circle
For the circle the path of minimum length is clearly the circumference. It's easy to prove that the minimum length therefore is less than $2\pi$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import glob
import imageio

plt.rcParams.update({'font.size': 18})

df = pd.read_csv("Best_L2.txt")
chromosomes = pd.read_csv("best_arrays.txt",header = None)

df_square = pd.read_csv("Best_L2_square.txt")
chromosomes_square = pd.read_csv("best_arrays_square.txt",header = None)

circ = pd.read_csv("circ_city.txt")
square = pd.read_csv("square_city.txt")

bh_circ = pd.read_csv("best_half_circ.txt")
bh_square = pd.read_csv("best_half_square.txt")
```

```python
df["L2"].min()
best = chromosomes.iloc[df["L2"].argmin()].to_numpy()
best = best -1

plt.figure(figsize=(6,6))
sns.scatterplot(circ,x="x",y="y",color = 'r',label="Cities")
plt.plot(circ["x"][0],circ["y"][0],marker="*", markersize = 7,label = "Start", linestyle = "")
plt.plot(circ["x"][best],circ["y"][best],color="g",label="Path")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.95))
```

```python
x = np.linspace(1, len(df),len(df))
plt.plot(x,df["L2"],label="best length")
plt.plot(x,bh_circ,color="r",label="best half average")
plt.axhline(y=2*np.pi, color = 'k', label = "Circumference length")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.95))
plt.xlabel("Generations")
plt.ylabel("$L^2$")
```

The genetic algorithm gives indeed the expected results. It's curious to see how the evolution occurs, the gif below provide and idea of this evolution.

## Square
In this case there isn't a theoretical approach to the minimum value, and we can't be sure that what we have found is the optimal path.

```python
df_square["L2"].min()
best_square = chromosomes_square.iloc[df_square["L2"].argmin()].to_numpy()
best_square = best_square -1

plt.figure(figsize=(6,6))
sns.scatterplot(square,x="x",y="y",color = 'r',label="Cities")
plt.plot(square["x"][0],square["y"][0],marker="*", markersize = 7,label = "Start", linestyle = "")
plt.plot(square["x"][best_square],square["y"][best_square],color="g",label="Path")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.95))
```

```python
x = np.linspace(1, len(df_square),len(df_square))
plt.plot(x,df_square["L2"],label="best length")
plt.plot(x,bh_square,color="r",label="best half average")
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.95))
plt.xlabel("Generations")
plt.ylabel("$L^2$")
```
