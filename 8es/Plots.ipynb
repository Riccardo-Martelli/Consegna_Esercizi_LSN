```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

plt.rcParams.update({'font.size': 18})

x = np.linspace(0,99,100)
df = pd.read_csv("modulus.txt")
H = pd.read_csv("H.txt")
res = pd.read_csv("valori.txt")
```

```python
plt.title("Probabiliy of the test function")
plt.errorbar(x,df["I"],yerr=df["err"],ecolor='r', label="p con Metropolis")
plt.xlabel("Blocchi")
plt.ylabel("$\int|\psi(x)|^2$dx")
plt.legend()
plt.show()
```

```python
plt.title("Expectation value of the Hamiltonian")
plt.errorbar(x,H["H"],yerr=H["err"],ecolor='r', label="p con Metropolis")
plt.xlabel("Blocchi")
plt.ylabel("$<\hat{H}>_T$")
plt.legend()
plt.show()
```
