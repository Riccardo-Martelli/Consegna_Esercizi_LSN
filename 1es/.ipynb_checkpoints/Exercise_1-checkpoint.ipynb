## Exercise 1.1 <a class="anchor" id="es1"></a>

The scope of this exercises is to test the GNC, and become familiar with the method of data blocking that will be used throughout the laboratory.
Briefly data blocking consists in averages of single values $x_i$ in N blocks to obtain an average given by $<x>=\frac{1}{N} \sum^N_{i=1} x_i$ with standard deviation $\sigma^2 = -<x>^2+<x^2>$.\
This method is useful because is like doing an experiment for each block not correlated to the former one, this statement requires a more detailed analysis that will be studied in deep in exercise 7.

The first task is to analyze the convergence of the integral $\int_0^1r\, dr$, given r asa pseudo-casual number

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import scipy.stats as stats


plt.rcParams.update({'font.size': 18})

df = pd.read_csv("./1.1/test.txt")

M=100000              # Total number of throws
N=100                 # Number of blocks
L=int(M/N)            # Number of throws in each block, please use for M a multiple of N

x = np.arange(N)+1
#x*=L

plt.figure(figsize=(12,8))
plt.title("Error of the mean")
plt.errorbar(x,df["mean"]-0.5,yerr=df["error_mean"])
plt.xlabel('Blocks')
plt.ylabel('<r>-1/2')
plt.grid(True)
```

The graph shows the values of 100 blocks with 100000 trows each. This graph clearly shows a convergence of the integral and a progressive reduction of the std, which for MC methods is tipically proportional to $N^{-1}$.\
Now let's check the convergence of the std given by $<\sigma^2> = \int_0^1(r-1/2)^2\,dr$ and r still pseudo causal.

```python
plt.figure(figsize=(12,8))
plt.title("Error on the std")
plt.errorbar(x,df["std"]-1/12.,yerr=df["error_std"])
plt.xlabel('Blocks')
plt.ylabel('<(r-0.5)^2>-1/12')
plt.grid(True)
plt.show()
```

Here the graph shows the convergence of sigma, using data blocking.

# Test del $\chi^2$
The former results are valid assuming the r in sampled between 0 and 1, this can be verified using the $\chi^2$ test.\
To execute the $\chi^2$ test one has to divide in M=100 sub intervals with N=10000 throws each an use the following formula $\chi^2 = \sum_{i=1}^M\frac{((N_i-N)/M)^2}{N/M}$

The results can be plotted with a histogram

```python
plt.title("Histogram $\chi^2$")
df = pd.read_csv("./1.1/chi.txt")
plt.xlabel("$\chi^2$")
sns.histplot(df["chi^2"])
print("Average value chi^2: ",df["chi^2"].mean())
```

This results shows that $\chi^2$ is around 100 and the std is low, compatible with whats in the wikipedia link.

## Exercise 1.2 <a class="anchor" id="es3"></a>

In this exercise we want to extend the GNC to gaussian, exponential and cauchy distribution. Then test the central limit theorem for this distribution.\
The sampling is done with the method of the inverse CDF.

To plot this we can use a progressive number of throws, namely N= 1,2,10,100 and sum the results with $S_N = \sum^N_{i=1}x_i$, where $x_i$ is the sampling of the given distribution

```python
# Dataframes per i vari lanci

df1 = pd.read_csv("./1.2/real1.txt")
df2 = pd.read_csv("./1.2/real2.txt")
df10 = pd.read_csv("./1.2/real10.txt")
df100 = pd.read_csv("./1.2/real100.txt")

# Plots per la distribuzione uniforme per i vari lanci
plt.figure(figsize=(10,8))
plt.title("Uniform")
sns.histplot(df1["Sum_uniform"],label="Throws: 1")
sns.histplot(df2["Sum_uniform"],label="Throws: 2")
sns.histplot(df10["Sum_uniform"],label="Throws: 10")
sns.histplot(df100["Sum_uniform"],label="Throws: 100")
plt.legend()
```

```python
# Plots per la distribuzione esponenziale per i vari lanci

plt.figure(figsize=(10,8))
plt.title("Exponential")
sns.histplot(df1["Sum_exponential"],label="Throws: 1")
sns.histplot(df2["Sum_exponential"],label="Throws: 2")
sns.histplot(df10["Sum_exponential"],label="Throws: 10")
sns.histplot(df100["Sum_exponential"],label="Throws: 100")
plt.legend()
```

```python
# Plots per la distribuzione cauchy per i vari lanci
limit = 50.0
df1 = df1[(df1["Sum_cauchy"]< limit) & (df1["Sum_cauchy"]> -limit)]
df2 = df2[(df2["Sum_cauchy"]< limit) & (df2["Sum_cauchy"]> -limit)]
df10 = df10[(df10["Sum_cauchy"]< limit) & (df10["Sum_cauchy"]> -limit)]
df100 = df100[(df100["Sum_cauchy"]< limit) & (df100["Sum_cauchy"]> -limit)]

plt.figure(figsize=(10,8))
plt.title("Cauchy")
sns.histplot(df1["Sum_cauchy"],label="Throws: 1")
sns.histplot(df2["Sum_cauchy"],label="Throws: 2")
sns.histplot(df10["Sum_cauchy"],label="Throws: 10")
sns.histplot(df100["Sum_cauchy"],label="Throws: 100")
plt.legend()
```

### TEST CLT <a class="anchor" id="test"></a>

```python
# Fit dei dati nel caso N = 100

mu = df100["Sum_uniform"].mean()
sigma = df100["Sum_uniform"].std()

xmin=np.min(df100["Sum_uniform"])
xmax=np.max(df100["Sum_uniform"])
x = np.linspace(xmin,xmax,101)
g=stats.norm(loc=mu, scale=sigma) 

plt.title("CLT uniform distribution")
plt.plot(x, g.pdf(x), label="CLT",color='r')
sns.histplot(df100["Sum_uniform"],label="Tries: 100",stat="density")
plt.legend()

#test per vedere la bontà del fit
set1 = df100["Sum_uniform"]
y1=np.sort(set1)
idx1 = [ i/len(y1) for i in range(len(y1))]

yy=set1.to_numpy()
s,p = stats.kstest(yy,lambda x: 
                   stats.norm.cdf
                   (x, loc=mu, scale=sigma))
print("===p-value===")
print(p)
```

Suppose a comparison value for the p-value as $\alpha = 0.05$, then p$ > \alpha$, therefore the null hypotesis cannot be dicarded

```python
mu = df100["Sum_exponential"].mean()
sigma = df100["Sum_exponential"].std()

xmin=np.min(df100["Sum_exponential"])
xmax=np.max(df100["Sum_exponential"])
x = np.linspace(xmin,xmax,101)
g=stats.norm(loc=mu, scale=sigma) 

plt.title("CLT exponential distribution")
plt.plot(x, g.pdf(x), label="CLT",color='r')
sns.histplot(df100["Sum_exponential"],label="Tries: 100",stat="density")
plt.legend()

#test per vedere la bontà del fit
set1 = df100["Sum_exponential"]
y1=np.sort(set1)
idx1 = [ i/len(y1) for i in range(len(y1))]

yy=set1.to_numpy()
s,p = stats.kstest(yy,lambda x: 
                   stats.norm.cdf
                   (x, loc=mu, scale=sigma))
print("===p-value===")
print(p)
```

For the exponential p$ > \alpha$ therefore in this case the null hypotesis can be rejected, however using $\alpha =0.01$ we cannot refuse the null Hypotesis. Maybe with more samples (i.e. 100 -> 1000) the fit could result in a more satisfying p-value.

```python
xmin=np.min(df100["Sum_cauchy"])
xmax=np.max(df100["Sum_cauchy"])
x = np.linspace(xmin,xmax,101)
mu = df100["Sum_cauchy"].mean()
sigma = df100["Sum_cauchy"].std()
g=stats.norm(mu,sigma) 

plt.title("CLT for Cauchy")
plt.plot(x, g.pdf(x), label="CLT",color='r')
sns.histplot(df100["Sum_cauchy"],label="Tries: 100",stat="density")
plt.legend()
#test per vedere la bontà del fit
set1 = df100["Sum_cauchy"]
y1=np.sort(set1)
idx1 = [ i/len(y1) for i in range(len(y1))]

yy=set1.to_numpy()
s,p = stats.kstest(yy,lambda x: 
                   stats.norm.cdf
                   (x, loc=mu, scale=sigma))
print("==p-value==")
print(p)
```

The p-value is null therefore the null hypotesis can be rejected

```python
xmin=np.min(df100["Sum_cauchy"])
xmax=np.max(df100["Sum_cauchy"])
x = np.linspace(xmin,xmax,202)
mu,gamma = stats.cauchy.fit(df100["Sum_cauchy"])
g=stats.cauchy(mu,gamma) 

plt.title("CLT for Cauchy")
plt.plot(x, g.pdf(x), label="CLT",color='r')
sns.histplot(df100["Sum_cauchy"],label="Tries: 100",stat="density")
plt.legend()

#test per vedere la bontà del fit
set1 = df100["Sum_cauchy"]
y1=np.sort(set1)
idx1 = [ i/len(y1) for i in range(len(y1))]

yy=set1.to_numpy()
s,p = stats.kstest(yy,lambda x: 
                   stats.cauchy.cdf
                   (x, loc=mu, scale=gamma))
print("==p-value==")
print(p)
```

The p-value is high enough that we cannot discard the null hypotesis of a Cauchy distribution, indeed as before p$ > \alpha$.\
Note that this was expected since Cauchy distribution has an infinite variance and the standard CLT doesn't apply in this case

## Exercise 1.3<a class="anchor" id="es3"></a>

In this exercise we apply the GNC to evaluate $\pi$ by simulating Buffon experiment. A needle with lenght L is thrown casually on a floor with parallel lines with distance d>L between them, the probability that the needle touches one of them is $P = \frac{2L}{d\pi} $, therefore pi can be obtained as $\pi=\frac{2L}{Pd}= \lim_{N_{th}\to \infty} \frac{2L N_{th}}{dN_{hit}} $, where $N_{th}$ is the number of throws and $N_{hit}$ is the number of times it hits a line

```python
# Dataframes per i vari lanci

df = pd.read_csv("./1.3/Buffon.txt")
df_er = pd.read_csv("./1.3/err.txt")

M=100000              # Total number of throws
N=100                 # Number of blocks
L=int(M/N)            # Number of throws in each block, please use for M a multiple of N

x = np.arange(len(df["PI"]))
#x*=L
```

```python
plt.axhline(y=np.pi, color = 'g', linestyle = '-', label = "$\pi$")
plt.errorbar(x,df["PI"],yerr=df_er["err_pi"],ecolor='r')
plt.title("Approxomation of $\pi$")
plt.xlabel("Blocks")
plt.ylabel("$\pi$")
```

The graph show that with this method we can obtain value close to pi with an error always smaller.
